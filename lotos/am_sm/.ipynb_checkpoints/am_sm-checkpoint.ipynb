# Notebook to obtain optimum damping and smoothing values:



```python
import numpy as np
import matplotlib.pyplot as plt
```

## Damping Parameters:

```python
%%sh
#Run this script to extract the relevant parameters from the log file
#Run for both the sequence

#for a in $(seq 0.0 0.25 4.0)
for a in $(seq 4.5 0.5 10.5)
do
    #i=$(printf "%2.2f" $a);
    i=$(printf "%2.1f" $a);
    echo $i
    cat ./r1/log/log_damp_$i.txt | awk '/avdv_p/{print $0}'>./r1/log/log_damp_extract_$i.txt
    cat ./r1/log/log_damp_$i.txt | awk '/dtot/{print $0}'>>./r1/log/log_damp_extract_$i.txt
done
```

```python
#Find the average model variance and data variance
#First sequence
MVP=[]
DVP=[]
MVS=[]
DVS=[]
dl1=np.arange(0.5,4.25,0.25) 
for dmp in dl1:
    
    fi=open('./r1/log/log_damp_extract_%2.2f.txt'%dmp,'r')
    data=fi.readlines()
    ct=0
    mvptot=0
    mvstot=0
    for line in data:
        if 'avdv' in line and ct<4:
            mvptot+=float(line.split()[2])
            mvstot+=float(line.split()[4])
            #print(dvptot/4.0,dvstot/4.0)
            ct+=1
    avmvp=mvptot/4.0
    avmvs=mvstot/4.0
    MVP.append(avmvp)
    MVS.append(avmvs)
    
    ######################################33
    dctp=0
    dcts=0
    for line in data:
        if 'dtot_p' in line:
            dctp+=1
            if dctp>0:
                avdvp=float(line.split()[3])
            
        if 'dtot_s' in line:
            dcts+=1
            if dcts>0:
                avdvs=float(line.split()[3])
    #print(sm,avdvp,avdvs)
    DVP.append(avdvp)
    DVS.append(avdvs)
print(MVS,DVS)
```

```python
#second sequence
dl2=np.arange(4.5,10.5,0.5) 
for dmp in dl2:
    
    fi=open('./r1/log//log_damp_extract_%2.1f.txt'%dmp,'r')
    data=fi.readlines()
    ct=0
    mvptot=0
    mvstot=0
    for line in data:
        if 'avdv' in line and ct<4:
            mvptot+=float(line.split()[2])
            mvstot+=float(line.split()[4])
            #print(dvptot/4.0,dvstot/4.0)
            ct+=1
    avmvp=mvptot/4.0
    avmvs=mvstot/4.0
    MVP.append(avmvp)
    MVS.append(avmvs)
    
    ######################################33
    dctp=0
    dcts=0
    for line in data:
        if 'dtot_p' in line:
            dctp+=1
            if dctp>0:
                avdvp=float(line.split()[3])
            
        if 'dtot_s' in line:
            dcts+=1
            if dcts>0:
                avdvs=float(line.split()[3])
    #print(sm,avdvp,avdvs)
    DVP.append(avdvp)
    DVS.append(avdvs)
print(MVS,DVS)
```

```python
#Plotting model variance vs data variance for Damping parameters
dl=np.concatenate((dl1, dl2), axis=0)
n1=4
n2=4
print("DAMP=",dl[n1],dl[n2])
plt.rcParams.update({'font.size': 20})
fig=plt.figure(figsize=(16,8))

plt.subplot(1, 2, 1)
min_val = np.min(MVP)
max_val = np.max(MVP)
scaled_MVP = (MVP - np.min(MVP)) / (np.max(MVP) - np.min(MVP))
scaled_DVP = (DVP - np.min(DVP)) / (np.max(DVP) - np.min(DVP))
plt.plot(scaled_MVP,scaled_DVP,'bo-',markersize=10)
plt.scatter(scaled_MVP[n1],scaled_DVP[n1],color='red',s=290)
plt.grid()
plt.xlabel('Normalized Model Variance',fontsize=20)
plt.ylabel('Normalized Data Variance',fontsize=20)


plt.subplot(1, 2, 2)
min_val = np.min(MVS)
max_val = np.max(MVS)
scaled_MVS = (MVS - np.min(MVS)) / (np.max(MVS) - np.min(MVS))
scaled_DVS = (DVS - np.min(DVS)) / (np.max(DVS) - np.min(DVS))
plt.plot(scaled_MVS,scaled_DVS,'bo-',markersize=10)
plt.scatter(scaled_MVS[n2],scaled_DVS[n2],color='red',s=290)
plt.grid()
plt.xlabel('Normalized Model Variance',fontsize=20)
plt.ylabel('Normalized Data Variance',fontsize=20)
#plt.savefig("optimum damping.png", dpi=300)
```
