```python
#Calculation of Mc and b-value
#Function to define RMS error
def mse(actual, predicted):
    actual = np.array(actual)
    predicted = np.array(predicted)
    differences = np.subtract(actual, predicted)
    squared_differences = np.square(differences)
    return squared_differences.sum()

#define cutoff
E=[]
C=np.arange(0.5,2.3,0.1)
for c in C:

    MR1=[]
    MR2=[]
    YV1=[]
    YV2=[]

    #Split the list
    for i in range(len(MR)):
        if MR[i]<c:
            MR1.append(MR[i])
            YV1.append(YV[i])
        else:
            MR2.append(MR[i])
            YV2.append(YV[i])
        
    #Fit a line with 0 slope for first part
    #b1=np.mean(np.array(YV1)-0*np.array(MR1)) #best fitting intercept 
    b1=1.97 #we can also use a defined intercept
    fit1=[b1+0*m for m in MR1] #predicted values

    #Misfit value for first part
    err1=(mse(fit1,YV1))
    #print(err1)

    #second part

    z2 = np.polyfit(MR2, YV2, 1, full=True)
    fit2=[z2[0][1]+z2[0][0]*m for m in MR2] 

    #error
    err2=z2[1][0]
    #print(err2)
    E.append(err1+err2)
min_ind=(E.index(min(E)))
print("Best Mc value=",C[min_ind])
```

```python
fig = plt.figure(figsize=(8,6))

plt.plot(C,E,"-o",color="red")
plt.xlabel("Mc value")
plt.ylabel("Squared residual sum")
plt.title("Misfit variation for different Mc values")
plt.savefig("mc_minimum_galan.png",dpi=300)
```

```python
#Fitting for best Mc value and obtain b-value

MR1=[]
MR2=[]
YV1=[]
YV2=[]

#optimum Mc value
c=1.5

#Split the list
for i in range(len(MR)):
    if MR[i]<c:
        MR1.append(MR[i])
        YV1.append(YV[i])
    else:
        MR2.append(MR[i])
        YV2.append(YV[i])
        
#Fit a line with 0 slope for first part
#b1=np.mean(np.array(YV1)-0*np.array(MR1)) #intercept
b1=1.97
fit1=[b1+0*m for m in MR1] #predicted values
print(b1)
#second part
z2 = np.polyfit(MR2, YV2, 1, full=True)
fit2=[z2[0][1]+z2[0][0]*m for m in MR2] 
slope=z2[0][0]
print("Obtained b-value=",-slope)
```

```python
fig = plt.figure(figsize=(8,6))
ax = plt.gca()

plt.scatter(MR,YV,s=10)
plt.plot(MR1,fit1,color="red")
plt.plot(MR2,fit2,color="orange")

plt.text(2.05, 1.70, 'b-value=1.9', fontsize = 14)
plt.text(2.05, 1.80, 'Mc=1.5', fontsize = 14)

plt.xlabel('M')
plt.ylabel('log(M)')
plt.savefig("b-value_galan.png",dpi=300)
```

```python
x = []
y = []
DATE=[]
fig = plt.figure(figsize=(10,8))
for i in range(len(x1)):
    dt=datetime.datetime(year[i],mon[i],day[i],hr[i],mm[i],int(sec[i]))
    DATE.append(str(dt))
ax = plt.gca()

def animation_func(i):
    x.append(x1[i])
    y.append(-y1[i])
    #colors.append(np.random.rand(1))
    #area = random.randint(0,30) * random.randint(0,30)
    plt.xlim(0,40)
    plt.ylim(-25,0)
    plt.scatter(x, y,c="red",s=200,linewidth=0.5,edgecolors='black')
    #plt.text(5,-10,"%s"%DATE[i])
    ax = plt.gca()
    ax.text(7, -2, "%s" %DATE[i],size=16,
            bbox={'facecolor': 'white', 'alpha': 1.0, 'pad': 5},
            ha="center")
    plt.xlabel("Distance(km)",fontsize=20)
    plt.ylabel("Depth(km)",fontsize=20)
    plt.tick_params(axis='both', labelsize=15)
    ax.set_aspect('equal')
anim = FuncAnimation(fig, animation_func, 
                          interval = 500)

from IPython import display
import matplotlib as mpl 
mpl.rcParams['animation.ffmpeg_path'] = r'/Users/sankha/ffmpeg'
# converting to an html5 video
video = anim.to_html5_video()
  
# embedding for the video
html = display.HTML(video)
  
# draw the animation
display.display(html)
plt.close()
```
