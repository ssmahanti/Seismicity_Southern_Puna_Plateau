```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("dem.xyz", sep='\s+', header=None)
```

```python
#print(df)
lat1 = -25.961
lon1 = -66.942

dl = np.sqrt((lat1-df[1])**2 + (lon1-df[0])**2)*111.1
print(len(dl))
idx =np.where(dl<1)
print(idx)
print(len(idx[0]))
#print(df[1][idx])
for i in range(len(idx[0])):
    j=idx[0][i]
    print( df[1][j],df[0][j],dl[j])
```

```python
df1 = pd.read_csv("focmec_data_center.csv")
#print(df1)
elev1 = []
for i in range(len(df1['Date'])):
    lat1 = df1['Latitude'][i]
    lon1 = df1['Longitude'][i]
    dep = df1['Depth(km)'][i]
    dl = np.sqrt((lat1-df[1])**2 + (lon1-df[0])**2)*111.1
    idx =np.where(dl<1)
    elevlist=[]
    for i in range(len(idx[0])):
        j=idx[0][i]
        elevlist.append(df[2][j])
    print(lat1, lon1, dep, np.average(elevlist))
    elev1.append(np.average(elevlist)+dep*1000)
```

```python
df1 = pd.read_csv("focmec_data_normal.csv")
#print(df1)
elev2= []
for i in range(len(df1['Date'])):
    lat1 = df1['Latitude'][i]
    lon1 = df1['Longitude'][i]
    dep = df1['Depth(km)'][i]
    dl = np.sqrt((lat1-df[1])**2 + (lon1-df[0])**2)*111.1
    idx =np.where(dl<1)
    elevlist=[]
    for i in range(len(idx[0])):
        j=idx[0][i]
        elevlist.append(df[2][j])
    print(lat1, lon1, dep, np.average(elevlist))
    elev2.append(np.average(elevlist)+dep*1000)
```

```python
#plt.hist(elev1)
idlist1 = np.arange(1,len(elev1)+1)
idlist2 = np.arange(1,len(elev2)+1)

plt.scatter(idlist1, elev1)
plt.scatter(idlist2, elev2)
plt.ylabel("Elevation (m)", fontsize=16)
plt.xlabel("Event number", fontsize=16)
plt.legend(["Cluster 1","Cluster 2"])
#plt.savefig("elevation_cluster_wdepth.png", dpi=300)
```
