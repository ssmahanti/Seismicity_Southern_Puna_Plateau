# HypoDD Analysis notebook

```python
#Importing relevant libraries
import numpy as np
import pandas as pd
from geopy.distance import geodesic
import matplotlib.pyplot as plt
import pygmt
```

## Steps for HypoDD


1. Run velest_hypodd.py to create the phasefile: phase_input_hypodd_v2.txt. Average station elevation is added to depth from velest to resemble the hypoDD setup.

2. Run selectphase.pl with phase_input_hypodd_v2.txt and event catalog to select phases for that catalog. This is not necessary for the same catalog.

3. Run ph2dt

4. Edit the hypoDD.inp as required. We need to shift the velocity model as velest model is w.r.t sealevel. But for hypoDD it is with respect to average station elevation


## 1. Checking event pair quality

```python
%%sh
cat event.dat | awk '{print $3,$4,$5,$10}'>temp_evelist.txt #lat,lon,depth,event id
```

```python
df=pd.read_csv("temp_evelist.txt",sep=" ",header=None)
latlist=df[0].values.tolist()
lonlist=df[1].values.tolist()
deplist=df[2].values.tolist()
idd=df[3].values.tolist()
```

```python
#Distribution of pairing 

fi=open("dt.ct",'r') 
data=fi.readlines()
DIST_P=[]
DT_P=[]
DIST_S=[]
DT_S=[]
for line in data:
    if line.split()[0]=="#":
        eve1_id=int(line.split()[1])
        eve2_id=int(line.split()[2])
        idx1=idd.index(eve1_id)
        idx2=idd.index(eve2_id)
        #Events
        dep1=deplist[idx1]    
        dep2=deplist[idx2]
        loc1=(latlist[idx1],lonlist[idx1])
        loc2=(latlist[idx2],lonlist[idx2])
        
        dx=geodesic(loc1,loc2).km 
        dz=deplist[idx1]-deplist[idx2]
        dist=np.sqrt(dx**2 + dz**2) #Distance
    else:
        tt1=float(line.split()[1])
        tt2=float(line.split()[2])
        phs=line.split()[4]
        dt=abs(tt1-tt2)  #Travel time difference
        if (phs=="P"):
            DIST_P.append(dist)
            DT_P.append(dt)
        else:
            DIST_S.append(dist)
            DT_S.append(dt)

            
```

```python
#Plotting
fig=plt.figure(figsize=(16,15))

plt.subplot(2,2,1)
plt.scatter(DIST_P,DT_P,s=1)
m=1/4.0  #Lowest P velocity allowed
xx=np.arange(0,100,0.1)
yy=m*xx+0.5
plt.plot(xx,yy,color="r")
plt.xlabel("Distance(km)",fontsize=15)
plt.ylabel("P_Travel_time_difference(sec)",fontsize=15)
plt.ylim([0,7])
plt.xlim([0,20])

plt.subplot(2,2,2)
plt.scatter(DIST_S,DT_S,s=1)
m=1/2.3 #Lowest S velocity allowed
xx=np.arange(0,100,0.1)
yy=m*xx+0.5
plt.plot(xx,yy,color="r")
plt.xlabel("Distance(km)",fontsize=15)
plt.ylabel("S_Travel_time_difference(sec)",fontsize=15)
plt.ylim([0,7])
plt.xlim([0,20])

plt.subplot(2,2,3)
plt.hist(DT_P,bins=100,color="r",edgecolor="black")
plt.ylabel("Counts",fontsize=15)
plt.xlabel("P_Travel_time_difference(sec)",fontsize=15)
plt.xlim([0,7])

plt.subplot(2,2,4)
plt.hist(DT_S,bins=100,color="r",edgecolor="black")
plt.ylabel("Counts",fontsize=15)
plt.xlabel("S_Travel_time_difference(sec)",fontsize=15)
plt.xlim([0,7])

#plt.savefig("./plots/hypodd_dtct_selection.png",dpi=300)
```

## 2. Residual comparison

```python
%%sh
cat hypodd_puna_v5.reloc | awk '{ if ( $24<=100 ) print $0 }' > eventlist_clustered_hypodd.txt  #Hypodd output
```

```python
#Event selected from velest catalog which were relocated with hypoDD to compare the resiual

evelist_vel=np.loadtxt("../velest/output/catalog_velest_puna_crust_v5.txt") #Velest catalog 
eve_idx_vel=evelist_vel[:,9]
evelist_hyp=np.loadtxt("./eventlist_clustered_hypodd.txt") #HypoDD output
eve_idx_hyp=evelist_hyp[:,0]
eve_list_new=eve_idx_vel.tolist()

idlist=[]
fo1=open("eventlist_clustered_velest.txt",'w') #Output file selecting events which were clustered

for i in range(len(eve_idx_hyp)):
    k=np.where(eve_idx_vel==eve_idx_hyp[i])
    #print(k)
    print(*evelist_vel[int(k[0])],file=fo1)
    eve_list_new.remove(eve_idx_vel[k])
#print(eve_list_new)

fo2=open("eventlist_nocluster.txt",'w')
for idx in eve_list_new:
    ii = np.argmin(np.abs(eve_idx_vel-idx))
    print(*evelist_vel[ii], file=fo2)
#Everything is printed as float but we do not care as we need only the residuals    
```

```python
#Plotting the residuals
fig=plt.figure(figsize=(18,12))

df_res_ini=pd.read_csv("eventlist_clustered_velest.txt",sep="\s+",header=None)[11] #HypoDD input residuals
df_res_final=pd.read_csv("eventlist_clustered_hypodd.txt",sep="\s+",header=None)[22] #HypoDD Output residuals
#print(len(df_res_final),len(df_res_ini))

#Input residuals
plt.subplot(2,2,1)
plt.hist(df_res_ini*100,bins=50,edgecolor="black")
plt.ylabel("Counts",fontsize=20)
plt.xlabel("Initial  Residual (ms)",fontsize=20)
plt.xlim([0,100])
plt.ylim([0,300])
plt.title("Residual before HypoDD run",fontsize=20)

#Output residuals
plt.subplot(2,2,2)
plt.hist(df_res_final*100,bins=50,edgecolor="black",color="red")
plt.ylabel("Counts",fontsize=20)
plt.xlabel("Final Residual (ms)",fontsize=20)
plt.xlim([0,10])
plt.ylim([0,300])
plt.title("Residual after HypoDD run",fontsize=20)
plt.savefig("./plots/hypodd_residual_reduction.png",dpi=300)
```

# Error estimation with resampling

```python
%%sh
echo " " > ./resampling/run_puna_all.reloc
for ((i=1;i<=200;i++))
do
    awk '{print $0}' ./resampling/output/run_puna_$i.reloc >> ./resampling/run_puna_all.reloc
done    
```

```python
df2=pd.read_csv("hypodd_puna_v5.reloc",sep="\s+",header=None)
fo2 = open("./resampling/puna_events_err_v5.txt",'w')
for id1 in range(len(df2[0])):
    lat=df2[1][id1]
    lon=df2[2][id1]
    dep=df2[3][id1]
    df3=pd.read_csv("./resampling/run_puna_all.reloc",sep="\s+",header=None)
    idx=np.where(df3[0]==df2[0][id1])
    laterrlist= abs([df3[1][i] for i in idx[0]]-lat)*111.1*1000 #m
    lonerrlist= abs([df3[2][i] for i in idx[0]]-lon)*111.1*1000 #m
    deperrlist= abs([df3[3][i] for i in idx[0]]-dep)*1000 #m
    laterr=round(np.mean(laterrlist),3)
    lonerr=round(np.mean(lonerrlist),3)
    deperr=round(np.mean(deperrlist),3)
    print(df2[0][id1],df2[1][id1],laterr,df2[2][id1],lonerr,df2[3][id1],deperr,file=fo2)
```

```python
%%sh
cat hypodd_puna_v5.reloc | awk '{print $11,$12,$13,$14,$15,$16,$2,$3,$4-3.1,$23}'> temp1.txt
cat ./resampling/puna_events_err_v5.txt | awk '{print $3,$5,$7}'>temp2.txt
paste temp1.txt temp2.txt > ./resampling/hypodd_catalog_err_unsorted.txt
rm temp1.txt temp2.txt
```
