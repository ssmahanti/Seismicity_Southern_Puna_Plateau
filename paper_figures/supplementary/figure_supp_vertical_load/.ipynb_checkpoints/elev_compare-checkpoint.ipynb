## This notebook compares vertical load between northern and southern segment of Southern Puna Plateau

```python
#Import libraies
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
#Exclude the swarm events to get only tectonic events:
df1 = pd.read_csv("galan_swarm_events_hypodd_v5.txt", sep='\s+', header=None)
df2 = pd.read_csv("events_achi.dat", sep='\s+', header=None)

fo1=open("events_achi_tectonic.dat",'w')
c=0
for i in range(len(df2[0])):
    
    k1=abs((df2[0][i]-df1[7]))
    k2=abs((df2[1][i]-df1[6]))
    if min(k1)>=0.0001 and min(k2)>=0.0001:
        c+=1
        print(df2[0][i],df2[1][i],df2[2][i],file=fo1)
print(c)
```

```python
#Obtain depth from surface: Depth+Elevation
#Averaged over nearby nodes
#Northern segment
df_dem = pd.read_csv("dem_north_resamp.xyz", sep='\s+', header=None) #DEM File prepared with prepare_dem.sh
df1 = pd.read_csv("events_achi_tectonic.dat",sep="\s+", header=None) #event list

elev_achi = []
c=0
for i in range(len(df1[0])):
    c+=1
    lat1 = df1[1][i]
    lon1 = df1[0][i]
    dep = df1[2][i]
    dl = np.sqrt((lat1-df_dem[1])**2 + (lon1-df_dem[0])**2)*111.1
    idx =np.where(dl<1)
    elevlist=[]
    for i in range(len(idx[0])):
        j=idx[0][i]
        elevlist.append(df_dem[2][j])
    print(c,lat1, lon1, dep, np.average(elevlist))
    elev_achi.append(np.average(elevlist)/1000+dep)
```

```python
#Obtain depth from surface: Depth+Elevation
#Averaged over nearby nodes
#Southern segment
df_dem = pd.read_csv("dem_south_resamp.xyz", sep='\s+', header=None)
df1 = pd.read_csv("events_culam.dat",sep="\s+", header=None)

elev_culam = []
c=0
for i in range(len(df1[0])):
    c+=1
    lat1 = df1[1][i]
    lon1 = df1[0][i]
    dep = df1[2][i]
    dl = np.sqrt((lat1-df_dem[1])**2 + (lon1-df_dem[0])**2)*111.1
    idx =np.where(dl<1)
    
    elevlist=[]
    for i in range(len(idx[0])):
        j=idx[0][i]      
        elevlist.append(df_dem[2][j])
    print(c,lat1, lon1, dep, np.average(elevlist))
    elev_culam.append(np.average(elevlist)/1000+dep)
```

```python
plt.rcParams.update({'font.size': 20})
fig=plt.figure(figsize=(16,6))
bins=np.arange(0,30)
plt.subplot(1, 2, 1)
plt.hist(elev_achi,bins,color='red',edgecolor='black')
plt.xlim([0,30])
plt.ylim([0,40])
plt.xlabel('Depth (km)')
plt.ylabel('Count')
plt.legend(['North'])
plt.subplot(1, 2, 2)
plt.hist(elev_culam,bins,color='green',edgecolor='black')
plt.xlim([0,30])
plt.ylim([0,40])
plt.xlabel('Depth (km)')
plt.ylabel('Count')
plt.legend(['South'])
plt.savefig("vertical_load_compare.jpg", dpi=300)
```
